```python
from mewpy.simulation import Environment

M9 = Environment.from_model(bu)
M9 = { k.replace('EX', 'R_EX'): v for k, v in M9.items() }
M9 = Environment.from_reactions(M9)
M9
```

```python
sample1 = CommunityModel([bt,bu,fn],flavor='reframed', add_compartments=True, merge_biomasses=True,abundances=[1,4,0.5])
sample2 = CommunityModel([bt,bu,fn],flavor='reframed', add_compartments=True, merge_biomasses=True)
sample3 = CommunityModel([bt,bu,ri],flavor='reframed', add_compartments=True, merge_biomasses=True)
sample4 = CommunityModel([bt,bu,sp],flavor='reframed', add_compartments=True, merge_biomasses=True)
sample5 = CommunityModel([bt,bu,ss],flavor='reframed', add_compartments=True, merge_biomasses=True)
```

```python
sim1 = sample1.get_community_model()
```

```python
sim1.set_environmental_conditions(M9)
```

```python
constraints = {'R_EX_fe2[e]':(-0.00007,0)}
solution1 = sim1.simulate(constraints=constraints)

print(solution1)
```

```python
from mewpy.cobra.com import SteadyComVA

l_va = np.linspace(0.1,1.0,num=10)

for va in l_va:
    va = round(va, 1)
    variability = SteadyComVA(sample1, obj_frac=va, constraints=M9)
    print(f'Strain\tMin\tMax\tVariability - {va}')
    for strain, (lower, upper) in variability.items():
        print(f'{strain}\t{lower:.1%}\t{upper:.1%}')
```
