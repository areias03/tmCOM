```python
from reframed.io.sbml import load_cbmodel
from mewpy.simulation import get_simulator
```

```python
bt = load_cbmodel('../models/non-ec/Bacteroides_thetaiotaomicron_VPI_5482.xml')
bu = load_cbmodel('../models/non-ec/Bacteroides_uniformis_ATCC_8492.xml')
ec = load_cbmodel('../models/non-ec/Escherichia_coli_ED1a.xml')
cc = load_cbmodel('../models/non-ec/Coprococcus_comes_ATCC_27758.xml')
fn = load_cbmodel('../models/non-ec/Fusobacterium_nucleatum_subsp_nucleatum_ATCC_25586.xml')
ri = load_cbmodel('../models/non-ec/Roseburia_intestinalis_L1_82.xml')
sp = load_cbmodel('../models/non-ec/Streptococcus_parasanguinis_ATCC_15912.xml')
ss = load_cbmodel('../models/non-ec/Streptococcus_salivarius_DSM_20560.xml')
```

```python
from mewpy.cobra.com import *

mets, rxns, over = jaccard_similarity_matrices([bt,bu,ec,cc,fn,ri,sp,ss])
```

```python
from mewpy.model import CommunityModel

sample1 = CommunityModel([bt,bu,ec],flavor='cobra')
```

```python
from reframed import FBA
from mewpy.simulation import Environment

M9 = Environment.from_model(ec)
super_oganism = sample1.merged_model
solution = FBA(super_oganism, constraints=M9)

print(solution)
solution.show_values(pattern='R_EX')
```

```python
sim1 = sample1.get_community_model()
```

```python
solution = sim1.simulate(constraints=M9)

print(solution)
solution.find('EX')
```
