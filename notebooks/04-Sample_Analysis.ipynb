```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Single models

## Media growth test

### Non constrained models

```python
ne_media_growth_test = pd.read_csv('../data/media_testing/ne_media_growth_test.csv')
ne_media_growth_test = ne_media_growth_test.set_index('Constraints')
ne_media_growth_test.columns = ['Bt', 'Bu', 'Ec', 'Fn', 'Ri', 'Sp', 'Ss']
ne_media_growth_test
```

```python
g = sns.clustermap(ne_media_growth_test, cmap = "vlag",row_cluster=False)
g.tick_params(axis='both', which='both', length=0)
g.ax_heatmap.set_ylabel('')
plt.setp(g.ax_heatmap.get_yticklabels(), rotation=0)
plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90)
plt.savefig('../data/images/ne_clustermap_mgt.png', bbox_inches='tight')
```

# Community models

## Non constrained samples

```python
ne_fba_all = pd.read_csv('../data/results/ne_fba.csv')
ne_fba_all = ne_fba_all.set_index('Constraints')
ne_fba_all.columns = ['Bt + Bu', 'Bt + Ec', 'Bt + Fn', 'Bt + Ri', 'Bt + Sp', 'Bt + Ss']
ne_fba_all
```

```python
g = sns.heatmap(ne_fba_all, cmap = 'vlag',annot=True)
g.tick_params(axis='both', which='both', length=0)
g.set_ylabel('')
plt.savefig('../data/images/ne_heatmap_fba_all.png', bbox_inches='tight')
```

```python
g = sns.clustermap(ne_fba_all, cmap = "vlag")
g.tick_params(axis='both', which='both', length=0)
g.ax_heatmap.set_ylabel('')
plt.setp(g.ax_heatmap.get_yticklabels(), rotation=0)  # For y axis
plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90) # For x axis
plt.savefig('../data/images/ne_clustermap_fba_all.png', bbox_inches='tight')
```

```python
ne_fba_single = pd.read_csv('../data/results/ne_fba_single.csv')
ne_fba_single = ne_fba_single.iloc[:,1:]
ne_fba_single
```

```python
ne_fba_single = ne_fba_single.drop(ne_fba_single[(ne_fba_single['Constraints'] == 'No constraints')].index)

g = sns.catplot(
    data=ne_fba_single, kind="bar",
    x="Partner", y="Growth", hue="Constraints",
    palette="dark", alpha=.6, height=5
)
g.despine(left=True)
g.set_axis_labels("", "Growth (gDW/h)")
g.set_xticklabels(['Bu','Ec','Fn','Ri','Sp','Ss'])
g.legend.set_title("Media")
plt.savefig('../data/images/ne_bt_growth_partners.png')
```

## Enzymatic constrained samples

### Flux Balance Analysis (FBA)

```python
ec_fba_all = pd.read_csv('../data/results/ec_fba_all.csv')
ec_fba_ecss = pd.read_csv('../data/results/ec_fba_ecss.csv')
ec_fba_ecss = ec_fba_ecss.iloc[:,1:]
ec_fba_all = pd.concat([ec_fba_all,ec_fba_ecss],axis=1)
ec_fba_all = ec_fba_all.set_index('Constraints')
ec_fba_all = ec_fba_all[sorted(ec_fba_all)]
ec_fba_all.columns=['Bt + Bu','Bt + Ec', 'Bt + Fn', 'Bt + Ri','Bt + Sp', 'Bt + Ss']
ec_fba_all
```

```python
g = sns.heatmap(ec_fba_all, cmap = 'vlag',annot=True)
g.tick_params(axis='both', which='both', length=0)
g.set_ylabel('')
plt.savefig('../data/images/ec_heatmap_fba_all.png', bbox_inches='tight')
```

```python
g = sns.clustermap(ec_fba_all, cmap = "vlag")
g.tick_params(axis='both', which='both', length=0)
g.ax_heatmap.set_ylabel('')
plt.setp(g.ax_heatmap.get_yticklabels(), rotation=0)  # For y axis
plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90) # For x axis
plt.savefig('../data/images/ec_clustermap_fba_all.png', bbox_inches='tight')
```

```python
ec_fba_single = pd.read_csv('../data/results/ec_fba_single_all.csv')
ec_fba_single_ecss = pd.read_csv('../data/results/ec_fba_single_ecss.csv')
ec_fba_single = pd.concat([ec_fba_single,ec_fba_single_ecss])
ec_fba_single = ec_fba_single.sort_values(by='Partner')
ec_fba_single = ec_fba_single.iloc[:,1:]
ec_fba_single
```

```python
ec_fba_single = ec_fba_single.drop(ec_fba_single[(ec_fba_single['Constraints'] == 'No constraints')].index)

g = sns.catplot(
    data=ec_fba_single, kind="bar",
    x="Partner", y="Growth", hue="Constraints",
    palette="dark", alpha=.6, height=5
)
g.despine(left=True)
g.set_axis_labels("", "Growth (gDW/h)")
g.set_xticklabels(['Bu','Ec','Fn','Ri','Sp','Ss'])
g.legend.set_title("Media")
plt.savefig('../data/images/ec_bt_growth_partners.png')
```
