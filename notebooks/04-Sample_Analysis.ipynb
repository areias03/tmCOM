```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Enzymatic constrained samples

```python
fba_all = pd.read_csv('../data/results/fba_all.csv')
fba_all = fba_all.set_index('Constraints')
fba_all_non = fba_all.replace(0,1)
log_fba_all = -(np.log10(fba_all_non))
fba_all
```

```python
log_fba_all
```

```python
x_labels = ['Bt + Bu','Bt + Ec ED1a','Bu + Ec ED1a','Bt + SS']
g = sns.heatmap(fba_all, xticklabels = x_labels, cmap = "vlag",annot=True)
g.tick_params(axis='both', which='both', length=0)
g.set_ylabel('')
plt.savefig('../data/images/ec_heatmap_fba_all.png', bbox_inches='tight')
plt.show()
```

```python
x_labels = ['Bt + Bu','Bt + Ec ED1a','Bu + Ec ED1a','Bt + SS']
g = sns.clustermap(fba_all, xticklabels = x_labels, cmap = "vlag")
g.tick_params(axis='both', which='both', length=0)
g.ax_heatmap.set_ylabel('')
plt.setp(g.ax_heatmap.get_yticklabels(), rotation=0)  # For y axis
plt.setp(g.ax_heatmap.get_xticklabels(), rotation=90) # For x axis
plt.savefig('../data/images/ec_clustermap_fba_all.png', bbox_inches='tight')
```
