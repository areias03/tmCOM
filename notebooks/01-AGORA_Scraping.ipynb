```python
import pandas as pd
import requests
```

```python
url = r'https://www.vmh.life/files/reconstructions/AGORA2/version2.01/sbml_files/individual_reconstructions'
tables = pd.read_html(url,flavor="html5lib") # Returns list of all tables on page
print(tables)
table = tables[0] # Select table of interest
```

```python
table = pd.DataFrame(table)
```

```python
table.head()
```

```python
table = table.drop(columns=["Unnamed: 0", "Last modified","Size","Description"])
```

```python
table = table.drop(table.index[[0,1,7304]])
table.rename(columns={"Name":"Filename"},inplace=True)
print(table)
```

```python
list_of_strains = pd.read_excel("../data/list_of _strains.xlsx")
```

```python
list_of_strains["Species"] = list_of_strains["Species"].str.replace(' ', '_')
```

```python
list_of_strains = list_of_strains.drop(columns=["Taxon_ID","Phylum","Class","Order","Family","Genus","Source","StrainID_database","On_AGORA"])
```

```python
list_of_strains
```

```python
file_list1 = list_of_strains["File_1"].to_list()
file_list2 = list_of_strains["File_2"].to_list()
file_list3 = list_of_strains["File_3"].to_list()
file_list4 = list_of_strains["File_4"].to_list()
file_list5 = list_of_strains["File_5"].to_list()

file_list = file_list1+file_list2+file_list3+file_list4+file_list5
file_list = [x for x in file_list if str(x) != 'nan']
```

```python
print(file_list)
```

```python
table_bool = table["Filename"].isin(file_list)
print(table_bool)
```

```python
table_bool[table_bool==True]

list_index = []
for item in table_bool[table_bool==True].index:
    list_index.append(item)

list_index = [x-2 for x in list_index]

print(list_index)
```

```python
match_list = table.iloc[list_index].iloc[:,0].to_list()
match_list[0]
```

```python
for match in match_list:
    url = f'https://www.vmh.life/files/reconstructions/AGORA2/version2.01/sbml_files/individual_reconstructions/{match}'
    response = requests.get(url)
    with open(f'../models/non-ec/agora2/{match}', 'w') as f:
        f.write(response.text)
```
