```python
from reframed.io.sbml import load_cbmodel
from cobra.io import read_sbml_model, write_sbml_model
from mewpy.cobra.util import add_enzyme_constraints
from mewpy.simulation import get_simulator
from mewpy.simulation.environment import Environment as Environment
from mewpy.util.request import retreive_gene,retreive_protein,brenda_query,get_smiles
import pandas as pd
import json
```

```python
model = read_sbml_model("../models/non-ec/iAF1260.xml")
```

```python
sim = get_simulator(model)
#sim.set_objective("Biomass")
```

```python
sim.simulate()
```

```python
data_all = dict()
try:
    with open(sim.id+".json", 'r') as f:
        data = json.load(f)
except:
    data = dict()
    for gene in sim.genes:
        try:
            info = retreive_gene(gene,"Escherichia coli")
            data[gene]={'protein':info['protein'],'sequence':info['seq'],'ec':[lm[1]for lm in info['properties']['Catalytic Activity']],'mw':info['mw']*0.001,'kcat':brenda_query(user = 'pg45962@uminho.pt',password='Mentafrio+15',ecNumber=[lm[1]for lm in info['properties']['Catalytic Activity']],organism = 'Escherichia coli')}
            data_all[gene]=info
        except:
            data[gene]={'protein': gene[len(sim._g_prefix):], 'mw': 0}
    with open(sim.id+".json", "w") as f:
        json.dump(data,f)
```

```python
print("Genes with missing data:")
for k,v in data.items():
    if v['mw']==0:
        print(k)
```

```python
df = pd.DataFrame.from_dict(data) 
df = df.transpose()
df = df.iloc[:,0:4]
df
```

## BRENDA Query

```python
for ec_l in df[['ec']]:
    for ec in ec_l:
        if ec:
            kcat = brenda_query('pg45962@uminho.pt','Mentafrio+15',ec,'Escherichia coli')
            if kcat:
                df[['kcat']] = kcat
```
