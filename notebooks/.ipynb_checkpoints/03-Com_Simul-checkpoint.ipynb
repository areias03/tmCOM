```python
from mewpy.simulation import Environment

M9 = Environment.from_model(ec)
M9
```

```python
M8 = Environment()

for r_id in simec.get_exchange_reactions():
    r_id2 = 'R_'+r_id
    M8[r_id2] = simec.get_reaction_bounds(r_id)

M8    
```

```python
sample1 = CommunityModel([bt,bu,ec,cc,fn,ri,sp,ss])
```

```python
sim1 = sample1.get_community_model()
```

```python
solution = sim1.simulate(constraints=M8)

print(solution)
```

```python
print(solution.find('R_EX').to_string())
```

```python
solution.find('R_biomass', sort=True,show_nulls=True)
```
